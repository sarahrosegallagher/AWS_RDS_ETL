```python
import pandas as pd
```

```python
df = pd.read_csv("vine_table.csv")
```

```python
df.head()
```

```python
df.count()
```

Step 1: new DataFrame or table to retrieve all the rows where the total_votes count is equal to or greater than 20

```python
total_votes_df = df[df.total_votes >= 20]
```

```python
#31k results 
total_votes_df
```

Step 2: retrieve all the rows where the number of helpful_votes divided by total_votes is equal to or greater than 50%.

```python
helpful_votes_df = total_votes_df[total_votes_df.helpful_votes/total_votes_df.total_votes >= 0.5]
```

```python
#28k results 
helpful_votes_df
```

Step 3: filter helpful_votes_df into two new dfs: one with paid (vine == "Y") and one with unpaid (vine == "N") reviews 

```python
paid_reviews_df = helpful_votes_df[helpful_votes_df.vine == "Y"]
```

```python
#61 results 
paid_reviews_df
```

```python
unpaid_reviews_df = helpful_votes_df[helpful_votes_df.vine == "N"]
```

```python
#28k results 
unpaid_reviews_df
```

Stats: total number of reviews, the number of 5-star reviews, and the percentage of 5-star reviews for the two types of review (paid vs unpaid).

```python
review_count = df.review_id.count()
```

```python
review_count
```

```python
five_star_count = df.star_rating[df.star_rating == 5].count()
```

```python
five_star_count
```

```python
paid_5_star = paid_reviews_df.star_rating[paid_reviews_df.star_rating == 5].count()

paid_count = paid_reviews_df.star_rating.count()

paid_5_star_ratio = paid_5_star/paid_count
```

```python
paid_5_star_ratio
```

```python
unpaid_5_star = unpaid_reviews_df.star_rating[unpaid_reviews_df.star_rating == 5].count()

unpaid_count = unpaid_reviews_df.star_rating.count()

unpaid_5_star_ratio = unpaid_5_star/unpaid_count
```

```python
unpaid_5_star_ratio
```

```python
review_stats_data = {"Total Review Count": [review_count],
                    "5 Star Review Count": [five_star_count],
                    "Paid 5 Star %":[paid_5_star_ratio],
                    "Unpaid 5 Star %": [unpaid_5_star_ratio]}
```

```python
reivew_stats_df = pd.DataFrame(review_stats_data)

```
